````python
# Define  fuction named get_stars and pass the variabel which are having star value and possing the div and class property#
````

````python
def get_stars(doc):
    star_all_tag = doc.find_all('div', class_='user_score_chart')
    stars =[]
    
    for star in star_all_tag:
        try:
            stars.append(star['data-percent'])
        except:
            stars.append(' ')
    return stars
    
````

## Defininig  the Helper Function 

````python
# Helper function is get_title and variable is doc is applying

def get_titles(doc):
    h2_tag_all = doc.find_all('h2')
    titles = []
# looping all the titles which are inside of the h2 tag and appending it's title
    
    for title in h2_tag_all[4:]:
        try:
            titles.append(title.text)
# if there in no title then it will come to exept function and give the value  None found            
        except:
            titles("None found")
            
# listing  all the title text  in new row after looping and appending each time
            
    return titles
             

    
````

## Creating list of releasing dates for movies and TV show

````python
# Defining helper function date_movie and pass beautifulSoup doc
def date_movie(doc):
    release_date = doc.find_all('div', class_='content')
    dates =[]

    for date in release_date:
        try:
            if not date.p.has_attr('class'):
                dates.append(date.p.text)
        except:
            print("None found")
    return dates
````

## All urls from the five pages are looping and Combining In One list of Dictionary 

````python
def all_details(urls):
    all_dict = { 'Title': [], 'Rating' : [], 'Release_Date' : []}
    for i in urls:
        response = requests.get(i)
        doc = BeautifulSoup(response.text)
        all_dict['Title'].extend(get_titles(doc))
        all_dict['Rating'].extend(get_stars(doc))
        all_dict['Release_Date'].extend(date_movie(doc))
        
    return all_dict
````
